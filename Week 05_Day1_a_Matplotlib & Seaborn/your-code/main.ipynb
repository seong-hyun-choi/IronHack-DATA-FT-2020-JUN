#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
fitbit.Months_encoded.unique()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
df_month=fitbit.groupby(by='Months_encoded')['Steps'].median()
df_month
```

```python
fitbit['Work_or_Weekend'].unique() # Work=1 & Weekend = 0
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
df_week=fitbit['Steps'].groupby(by=fitbit['Work_or_Weekend']).median()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,4))

axes[0] = df_month.plot(kind='line', ax=axes[0] ,lw=3, c='green', marker='o')
axes[1] = df_week.plot(kind='bar', ax=axes[1], color=list('rb'))

#ax.set_title
axes[0].set_title('Median Steps by Month', fontweight='bold', fontsize=15)
axes[1].set_title('Median Steps by Workday and Weekend', fontweight='bold', fontsize=15)

#ax.set_xticks
axes[0].set_xticks(np.arange(1,13))
axes[1].set_xticks([0,1])

#ax.set_xticklabels
axes[0].set_xticklabels(['Jan','Fev','Mar','Abr','May','Jun', 'Jul','Ago', 'Sep','Oct','Nov', 'Dez'])
axes[1].set_xticklabels(['Weekend','Work'], rotation='horizontal')

#ax.set_ylim
axes[0].set_ylim([0,15000])
axes[1].set_ylim([0,15000])

#ax.grid()
axes[0].grid(alpha=0.3)
axes[1].grid(alpha=0.3)

#ax.set_xlabel
axes[0].set_xlabel('MONTH', fontweight='bold')
axes[1].set_xlabel('WEEKEND / WORK', fontweight='bold')

#ax.set_ylabel
axes[0].set_ylabel('MEDIAN NUMBER OF STEPS', fontweight='bold')
axes[1].set_ylabel('MEDIAN NUMBER OF STEPS', fontweight='bold')

#ax.set_facecolor (graph backgrond color)
axes[0].set_facecolor('#C6F7F7') 
axes[1].set_facecolor('#F2B3EA')

# fig background color
fig.patch.set_facecolor('#CFD1D0')

plt.savefig('figure/graph_steps.png', dpi=300)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
df= fitbit.loc[:,['Minutes Lightly Active','Minutes Very Active', 'Minutes Sedentary','Steps']]
df.head()
```

```python
fig,axes = plt.subplots(1,3,figsize=(15,6))

axes[0].scatter(df.iloc[:,0], df['Steps'])
axes[1].scatter(df.iloc[:,1], df['Steps'])
axes[2].scatter(df.iloc[:,2], df['Steps'])

# to make a loop

lst_color=['r','b','g']
lst_graph_c = ['#DEFAFA','#EEF2D5','#F5E4F7']
for i in range(3):
    axes[i].scatter(df.iloc[:,i], df['Steps'], color=lst_color[i])
    axes[i].set_xlabel(df.iloc[:,i].name, fontweight='bold', fontsize=12)
    axes[i].set_ylabel('Number of Steps', fontweight='bold', fontsize=12)
      
    axes[i].set_facecolor(lst_graph_c[i])
    axes[i].grid()
```
