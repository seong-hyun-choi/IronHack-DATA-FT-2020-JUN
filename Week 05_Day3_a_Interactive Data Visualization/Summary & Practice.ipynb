## Bar chart: df. iplot ( kind='bar', x= column) 

### df.column.apply (lambda x: 1 if x== 'Yes' else 0)

````python
# create variable ChurnBinary (1-0) based on Churn (Yes-No)
df['ChurnBinary'] = df['Churn'].apply(lambda x : 1 if x == 'Yes' else 0)
df.head()
````

````python
df.PaymentMethod.unique()
````

````python
# get the average of churn for each payment method
data = df.groupby('PaymentMethod', as_index=False)['ChurnBinary'].mean()
data
````

````python
data.plot(kind='bar', x='PaymentMethod')
````

````python
# Interactive bar chart
data.iplot(kind='bar', x='PaymentMethod')
````

# IPyWidgets : from ipywidgets import interact

````python
from ipywidgets import interact
````

````python
df['TotalCharges'].iplot(kind='hist', bins=100, title='Tenure Distribution')
````

## def plot_graph(bin)

````python
def plot_graph(bins):
    df['TotalCharges'].iplot(kind='hist', bins=bins, title='Tenure Distribution')
````

## interact(function_name, arguments= ( , , ))

````python
interact(plot_graph, bins=(0,100,10)) # (start, stop, step)
````

## @interact(arguments=( , , ))  / function {df.iplot()}

````python
@interact(bins = (5, 50, 5))
def plot_graph(bins):
    df['TotalCharges'].iplot(kind='hist', bins=bins, title='Tenure Distribution')
````

##  @interact(arguments=[ , ]) / function { df.query / df.iplot }

````python
@interact(churn=[True, False])
def plot_graph(churn):
    data =df.query(f'ChurnBinary == {churn}')
    
    data['tenure'].iplot(kind='hist', bins=50, title='Tenure Distribution')
````

````python
@interact(Gender=list(df['gender'].unique()), 
          Partner=list(df['Partner'].unique()),
          Internet=list(df['InternetService'].unique()), 
          Payment=list(df['PaymentMethod'].unique())
         )

def scatter(Gender, Partner, Internet, Payment):
    
    data = df.loc[(df['gender']==Gender) & 
                  (df['Partner']==Partner) & 
                  (df['InternetService']==Internet) & 
                  (df['PaymentMethod']==Payment), :]

    data.iplot(kind='scatter', x='tenure', y='MonthlyCharges', 
               categories='Contract', text='customerID', 
               xTitle='Tenure', yTitle='Monthly Charges',
               title='Charges vs. Tenure')
````

# Interactivity is not limited to plots

One can use the `display` function to show dataframes interactively.

## from IPython.display import display

````python
from IPython.display import display

display(df.head())
display(df.head(3))
````

````python
@interact(gender=list(df['gender'].unique()),  
          payment=list(df['PaymentMethod'].unique()))
         
def report(gender, payment):
    condition = (df.gender == gender) & (df.PaymentMethod == payment)
    data = df.loc[condition, :]
    
    display(data.groupby('Churn').mean())
    
````

## Pandas Styling :  .display(  df.style.highlight_max() )

````python
from IPython.display import display

@interact(payment=['Electronic check', 'Mailed check'])  

def report(payment):
    condition = df.PaymentMethod == payment
    data = df.loc[condition, :]
    
    display(data.groupby('Churn').mean().style.highlight_max())
````
